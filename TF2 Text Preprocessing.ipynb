```
# More Imports
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
# Just a simple test
sentences = [

    "I like eggs and ham.",
    "I love chocolate and bunnies.",
    "I hate onions."         
    
]
```

```
MAX_VOCAB_SIZE = 20000
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE)
tokenizer.fit_on_texts(sentences)
sequence = tokenizer.texts_to_sequences(sentences)
```

```
print(sequence)
```

```
# How to get the word to index mapping?
tokenizer.word_index
```

```
# use the defaults
data = pad_sequences(sequence)
print(data)
```

```
MAX_SEQUENCE_LEGNTH = 5
data = pad_sequences(sequence, maxlen=MAX_SEQUENCE_LEGNTH)
print(data)
```

```
data = pad_sequences(sequence, maxlen=MAX_SEQUENCE_LEGNTH, padding='post')
print(data)
```

```
# too much padding
data = pad_sequences(sequence, maxlen=6)
print(data)
```

```
data = pad_sequences(sequence, maxlen=4)
print(data)
```

```
data = pad_sequences(sequence, maxlen=4, truncating='post')
print(data)
```
